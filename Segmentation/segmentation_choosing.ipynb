Constant and redundant features were removed prior to clustering, as they do not contribute to distance-based separation and may distort clustering results.

```python
cols_to_drop = [
    'order_frequency',
    'min_items_per_order',
    'active_days_count'
]

clustering_df = clustering_df.drop(columns=cols_to_drop)
```

```python
clustering_df.describe().T[['mean', 'std', 'min', 'max']]
```

```python
clustering_df.to_csv("clustering_dataset_final.csv", index=False)
```

Since clustering algorithms are distance-based, features with larger numeric ranges may dominate the clustering process. Therefore, all features were standardized to ensure equal contribution.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(clustering_df)

X_scaled_df = pd.DataFrame(
    X_scaled,
    columns=clustering_df.columns
)

X_scaled_df.head()
```

```python
X_scaled_df.describe().T[['mean', 'std']]
```

```python
X_scaled_df.to_csv("clustering_dataset_scaled.csv", index=False)
```

In addition to standard normalization, RobustScaler was tested as it is less sensitive to outliers and scales features based on median and interquartile range.

```python
from sklearn.preprocessing import RobustScaler

robust_scaler = RobustScaler()

X_robust = robust_scaler.fit_transform(clustering_df)

X_robust_df = pd.DataFrame(
    X_robust,
    columns=clustering_df.columns
)

X_robust_df.head()
```

```python
X_robust_df.describe().T[['mean', 'std', 'min', 'max']]
```

```python
X_robust_df.to_csv("clustering_dataset_robust_scaled.csv", index=False)
```

```python
comparison = pd.DataFrame({
    'std_scaled_std': X_scaled_df.std(),
    'robust_scaled_std': X_robust_df.std()
})

comparison.sort_values('std_scaled_std', ascending=False).head(10)
```
