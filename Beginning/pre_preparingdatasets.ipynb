```python
import os
import pandas as pd
```

```python
path = r"C:\Diploma\Datasets\InstaCart Online Grocery Basket"
```

```python
aisles = pd.read_csv(path + r"\aisles.csv")
departments = pd.read_csv(path + r"\departments.csv")
products = pd.read_csv(path + r"\products.csv")
orders = pd.read_csv(path + r"\orders.csv")
prior = pd.read_csv(path + r"\order_products__prior.csv")
train = pd.read_csv(path + r"\order_products__train.csv")
```

```python
order_products = pd.concat([prior, train], ignore_index=True)
```

```python
df = order_products.merge(orders, on='order_id', how='left')
```

```python
df = df.merge(products, on='product_id', how='left')
```

```python
df = df.merge(aisles, on='aisle_id', how='left')
df = df.merge(departments, on='department_id', how='left')
```

```python
df.to_csv(r"C:\Diploma\Datasets\instacart_full.csv", index=False)
```
